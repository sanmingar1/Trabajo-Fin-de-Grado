```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
df = pd.read_csv('dataset_formateado.csv')
```

```python
df.head()
```

```python
#Eliminamos las entradas del dataset que hayan sido devoluciones
df = df[df['Sale or Return'] == 'sale']
```

```python
df['Date'] = df['Date'].astype(str)
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
df = df.drop('Date',axis=1)
df = df.drop('Time',axis=1)
```

```python
df.head()
```

```python
df['transaction_id'] =  df['Datetime'].dt.floor('min').factorize()[0] + 1
# df.head()
df = df[df['transaction_id'].map(df['transaction_id'].value_counts()) > 1]
df.head()
```

```python
product_matrix = pd.crosstab(df['transaction_id'], df['Item Code'])

# Usando el algoritmo de k-medias, encontrar los clusters de productos que se venden juntos para como colocar los productos
# en la tienda de forma que se vendan más.
```

```python
product_matrix
```

```python
kmeans = KMeans(n_clusters=15,random_state=42)
product_matrix['Cluster'] = kmeans.fit_predict(product_matrix)
product_matrix.head()
# cluster_labels = kmeans.labels_
```

```python
-----------------------------------------------------------------------------
# print(product_matrix.iloc[0].values)
co_occurrence_matrix = product_matrix.T.dot(product_matrix)
```

```python
print(co_occurrence_matrix.iloc[0].values)
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
scaler = cosine_similarity(co_occurrence_matrix)

# data_matrix_scaled = scaler(co_occurrence_matrix)

# Elegimos un número de clústers, Regla del codo
k = 13
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(scaler)

# Obtener las etiquetas de los clústers para cada producto/transacción
labels = kmeans.labels_
```

```python
# To visualize the K-means clustering, let's proceed with a PCA for dimensionality reduction,
# then plot the clusters in a 2D space. I'll perform the visualization using the data prepared in the notebook.

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns


# Step 1: Apply PCA to reduce dimensions to 2D for visualization
pca = PCA(n_components=2)
data_pca = pca.fit_transform(scaler)

# Step 2: Convert the PCA result to a DataFrame for easier plotting and add the cluster labels
df_pca = pd.DataFrame(data_pca, columns=['PCA1', 'PCA2'])
df_pca['cluster'] = labels  # Adding cluster labels to the dataframe

# Step 3: Visualize the clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='cluster', palette='viridis', s=60, alpha=0.7)
plt.title('Visualization of K-means Clusters on Products using PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```
